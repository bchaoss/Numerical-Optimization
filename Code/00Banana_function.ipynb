## Rosenbrock function
A non-convex mathematical function commonly used as a test problem for optimization algorithms:

$$f(x, y) = (a - x)^2 + b(y - x^2)^2$$

The following code defines the standard Rosenbrock function and plots its 3D surface and 2D contour plot. We will use this function multiple times in our notebooks to evaluate the performance of optimization algorithms.

```python
import numpy as np
import matplotlib.pyplot as plt

# define the Rosenbrock function
def Rosenbrock(x1, x2):
  a = 1
  b = 100
  return (a - x1)**2 + b * (x2 - x1*x1)**2

# Generate data for plotting
x1 = np.linspace(-1.1, 1.1, 100)
x2 = np.linspace(-2, 4, 400)
X1, X2 = np.meshgrid(x1, x2)
```

```python
# Plot
fig = plt.figure(figsize=(8, 4))
ax1 = plt.subplot(121, projection='3d')
ax2 = plt.subplot(122)

f = Rosenbrock
ax1.plot_surface(X1, X2, f(X1, X2), alpha=0.4, cmap='winter')
ax2.contour(X1, X2, f(X1, X2), cmap="winter")

# marking the global minimum point of the Rosenbrock function
ax1.plot(1, 1, 'o')
ax2.plot(1, 1, 'o')

ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('f')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
plt.suptitle('Rosenbrock Function', fontsize=12, fontweight='bold')
plt.show()
```
