Rosenbrock function 图像

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-1.1, 1.1, 100)
x2 = np.linspace(-2, 4, 400)
X1, X2 = np.meshgrid(x1, x2)


def f(x1, x2):
    return (1-x1)**2 + 100*(x2-x1*x1)**2


fig = plt.figure()
ax1 = plt.subplot(121, projection='3d')
ax2 = plt.subplot(122)
ax1.plot_surface(X1, X2, f(X1, X2), alpha=0.4, cmap='winter')
ax2.contour(X1, X2, f(X1, X2), cmap="winter")
ax1.plot(1, 1, 'o')
ax2.plot(1, 1, 'o')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('f')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
plt.show()
```
